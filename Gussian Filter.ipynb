```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread('im_q1.tif')
```

```python
def guassian(w,sd):
    return (1 / (np.sqrt(2 * np.pi)*sd) * np.e**(-(w**2)/ 2 * sd**2))
```

```python
def my2Dgussian(w,sigma=1):
    g_1D = np.linspace(-(w // 2),w // 2,w)
    for i in range(w):
        g_1D[i] = guassian(g_1D[i] ,sigma)
    g_2D = np.outer(g_1D,g_1D)
    g_2D = g_2D*100
    g_2D = g_2D.astype(int)
    value = np.sum(g_2D)
    g_2D = g_2D * 1.0/value
    return g_2D
```

```python
guassian1 = my2Dgussian(9,1)
guassian2 = my2Dgussian(17,1)
guassian3 = my2Dgussian(31,1)
output1 = cv2.filter2D(src = np.array(img1), ddepth=-1 , kernel= guassian1 )
output2 = cv2.filter2D(src = np.array(img1), ddepth=-1 , kernel= guassian2 )
output3 = cv2.filter2D(src = np.array(img1), ddepth=-1 , kernel= guassian3 )
```

```python
plt.figure(dpi=150)
plt.subplot(2, 2, 1)
plt.imshow(img1)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(output1)
plt.title('9-Guassian')
plt.axis('off')


plt.subplot(2, 2, 3)
plt.imshow(output2)
plt.title('17-Guassian')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(output3)
plt.title('31-Guassian')
plt.axis('off')
```
