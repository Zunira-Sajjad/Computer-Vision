```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img= cv2.imread('im_q3.tif')
```

```python
mask1 = np.array([[0,-1,0],
                    [-1,4,-1],
                    [0,-1,0]])
mask3 = np.array([[-1,-1,-1],
                    [-1,8,-1],
                    [-1,-1,-1]])
mask2 = np.array([[1,0,-1],
                    [2,0,-2],
                    [1,0,-1]])
```

```python
output = cv2.filter2D(src =np.array(img), ddepth=-1, kernel=mask1)
```

```python
output1 = cv2.filter2D(src = np.array(img), ddepth=-1, kernel=mask3)
```

```python
sharp_image1 = cv2.filter2D(src = np.array(output) , ddepth=-1, kernel=mask2)
```

```python
sharp_image2 = cv2.filter2D(src = np.array(output1) , ddepth=-1, kernel=mask2)
```

```python
plt.figure(dpi=150)


plt.subplot(2, 2, 1)
plt.imshow(output )
plt.title('90 Isotropic Image')
plt.axis('off')


plt.subplot(2, 2, 2)
plt.imshow(output1)
plt.title('45 Isotropic sharp Image')
plt.axis('off')


plt.subplot(2, 2,3 )
plt.imshow(sharp_image1)
plt.title('90 Isotropic sharp Image')
plt.axis('off')

plt.subplot(2, 2,  4 )
plt.imshow(sharp_image2)
plt.title('45 Isotropic sharp Image')
plt.axis('off')
```
